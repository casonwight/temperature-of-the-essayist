```python
import pandas as pd
import numpy as np
```

```python
prompts = pd.read_csv("autocomplete-results.csv").drop(columns=['Unnamed: 0'])
```

```python
i = 0
essay_output = pd.DataFrame(columns=['essay', 'full_prompt', 'short_prompt', 'act_output', 'temp', 'pred_output'])
for i in range(len(prompts)):
    essay = eval(prompts.loc[i, 'essay'])
    full_prompt = eval(prompts.loc[i, 'full_prompt'])
    short_prompt = eval(prompts.loc[i, 'short_prompt'])
    act_output = eval(prompts.loc[i, 'act_output'])
    pred_output = eval(prompts.loc[i, 'pred_output'])
    temp = np.array([x for x in prompts.loc[i, 'temp'].replace('[', '').replace(']', '').replace('\n', '').split(' ') if x]).astype(float)
    essay_output = essay_output.append(pd.DataFrame({
        'essay': essay,
        'full_prompt': full_prompt,
        'short_prompt': short_prompt,
        'act_output': act_output,
        'temp': temp,
        'pred_output': pred_output
    }), ignore_index=True)

essay_output
```

```python
essay_output = essay_output.append(pd.read_csv("autocomplete-results-3.csv").drop(columns=["Unnamed: 0"]).assign(essay=4), ignore_index=True)
essay_output
```

```python
essay_output.to_csv("autocomplete-results.csv")
```
